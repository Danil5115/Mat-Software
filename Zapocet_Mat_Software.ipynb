# Vizualizace dat

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Načtení dat z CSV souboru
titanic_data = pd.read_csv('train.csv') #Vytvoří se pohodlný soubor, odkud můžete brát data
```

```python
# 1 graf
sns.kdeplot(data=titanic_data, x="Age", hue="Survived", fill=True, alpha=0.4)
plt.title("Plošné rozdělení věku mezi přeživšími a nepřeživšími pasažéry")
plt.show()
```

```python
# 2 graf
# Vytvoření sloupcového grafu
sns.countplot(x='Pclass', hue='Survived', data=titanic_data)

# Nastavení popisků os a titulku grafu
plt.xlabel('Třída')
plt.ylabel('Počet')
plt.title('Přeživší vs. mrtví podle třídy')
plt.legend(['Mrtví', 'Přeživší'])

# Zobrazení grafu
plt.show()
```

```python
# 3 graf
# Definice trénovacích dat
sns.scatterplot(x='Age', y='Fare', data=titanic_data)
plt.title('Korelace mezi věkem a cenou jízdenky')
plt.show()
```

```python
#4 graf
# Vytvoření tabulky kontingenčních vztahů
ct = pd.crosstab([titanic_data['Pclass'], titanic_data['Sex']], titanic_data['Survived']) #принимает два или более факторов и возвращает таблицу связей между ними. vezme dva nebo více faktorů a vrátí tabulku vztahů mezi nimi.

# Nastavení velikosti grafu
plt.figure(figsize=(8, 6))

# Vykreslení heatmapy
sns.heatmap(ct, annot=True, fmt='d', cmap='Blues') #tabulka s daty pro vizualizaci, přidá číselné hodnoty do teplotní mapy, formát pro číselné hodnoty, colormap, definuje barevný gamut pro heatmapu

# Přidání titulku a popisků os
plt.title('Korelace mezi třídou, pohlavím a přežitím')
plt.xlabel('Přežil')
plt.ylabel('Třída a pohlaví')

plt.show()
```

```python
# 5 graf
# Počítání počtu přeživších a nepřeživších cestujících
survival_counts = titanic_data['Survived'].value_counts()

# Vytváření koláčového grafu
labels = ['Přeživší', 'Nepřeživší']
colors = ['#66CDAA', '#FF7F50']
explode = [0.1, 0] #расстояние между частями vzdálenost mezi díly
plt.pie(survival_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90, explode=explode) #формат написание числа autopct  číselný formát
plt.title('Rozdělení přeživších a nepřeživších cestujících')
plt.axis('equal')
plt.show()
```

```python
# 6 graf
# Nastavení velikosti grafu
plt.figure(figsize=(8, 6)) # размер графика в дюймах velikost grafu v palcích

# Vykreslení histogramu
plt.hist(titanic_data['Fare'], bins=30, color='purple') #bins количество столбцов sloupce Berou se údaje o cenách a cestujících

# Přidání titulku a popisků os
plt.title('Distribuce cen jízdenek')
plt.xlabel('Cena jízdenky')
plt.ylabel('Počet cestujících')

plt.show()
```

```python
#7 graf
sns.set_style('whitegrid') #стиль белая сетка styl bílé mřížky
sns.violinplot(x='Pclass', y='Age', hue='Sex', data=titanic_data, split=True, inner='quart')#inner='quart' zobrazuje pouze kvartily (25 %, 50 %, 75 %) v rámci jednotlivých houslí, což pomáhá vyhodnotit tvar distribuce dat.
plt.title('Rozdělení věku a třídy podle pohlaví')
plt.xlabel('Třída')
plt.ylabel('Věk')
plt.show()
```
